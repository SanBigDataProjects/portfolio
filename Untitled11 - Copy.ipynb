```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
data=pd.read_csv('normal_skew.csv')
data
```

```python
stats.probplot(data['normal_example'])
plt.show()
```

```python
stats.probplot(data['skewed_example'])
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 6))


# Q-Q plot for normal_example
stats.probplot(data['normal_example'], dist="norm", plot=ax[0])
#ax[0].set_title('Q-Q plot for normal_example')

# Q-Q plot for skewed_example
stats.probplot(data['skewed_example'], dist="norm", plot=ax[1])
#ax[1].set_title('Q-Q plot for skewed_example')

# Show the plots
plt.show()
```
